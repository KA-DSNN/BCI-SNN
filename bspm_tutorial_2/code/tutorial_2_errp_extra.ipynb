Epoching on Cz (Reference channel) in order to print the error related peak at 285.9 ms. The epoching was done between 0.5 seconds before the event and 1 second after the events.

```python
(events_from_annot,event_dict) = mne.events_from_annotations(eeg)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
fig, axs = plt.subplots(6,5, facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .5, wspace=.001)
axs = axs.ravel()
for i in range(len(eeg.ch_names)):
    epochs_noError = mne.Epochs(eeg,events_from_annot,tmin=-0.5,tmax=1.0,event_id=7,preload=True) # S 4
    epochs_Error = mne.Epochs(eeg,events_from_annot,tmin=-0.5,tmax=1.0,event_id=8,preload=True) # S 5
    epochs_noError.pick_channels([eeg.ch_names[i]])
    epochs_Error.pick_channels([eeg.ch_names[i]])
    axs[i].set_title(eeg.ch_names[i])
    ev_avg_noError = epochs_noError.average()
    ev_avg_Error = epochs_Error.average()
    axs[i].plot(ev_avg_noError.times[:],ev_avg_noError.data[0,:])
    axs[i].plot(ev_avg_Error.times[:],ev_avg_Error.data[0,:])
    axs[i].set_yticklabels([])
    axs[i].set_xticklabels([])
    axs[i].axvline(0.2, ls='--',c='green')
    axs[i].axvline(0.6, ls='--',c='green')
    #plt.legend(["No Error","Error"])
fig.delaxes(axs[27]) #The indexing is zero-based here
fig.delaxes(axs[28])
fig.delaxes(axs[29])
fig.show()
```
